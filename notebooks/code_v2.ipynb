```python
import pandas as pd
data = pd.read_excel('MOCK_Data_v2.xlsx')
data.head()
```

```python
# Clean text column (important)
data["Payment Delay"] = data["Payment Delay"].str.lower().str.strip()

# Step 1: Default = Low
data["Churn Risk"] = "Low"

# Step 2: Medium risk condition
data.loc[
    (
        (data["Usage Drop"] > 0.15) &
        (data["Number of Support Tickets"] > 2)
    ),
    "Churn Risk"
] = "Medium"

# Step 3: High risk condition (more strict)
data.loc[
    (
        (data["Usage Drop"] > 0.30) &
        (data["Number of Support Tickets"] > 3) &
        (data["Payment Delay"] == "yes")
    ),
    "Churn Risk"
] = "High"
```
